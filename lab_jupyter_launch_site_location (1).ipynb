```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# --- 1. Define Launch Site Coordinate (CCAFS SLC-40) ---
launch_site_lat = 28.563197
launch_site_lon = -80.576820
launch_site_coord = [launch_site_lat, launch_site_lon]

# --- 2. Define Proximity Coordinates (Found using MousePosition on the map) ---

# Closest City: Titusville, FL (approximate)
city_lat = 28.5735
city_lon = -80.8010

# Closest Railway Track (approximate)
railway_lat = 28.57215
railway_lon = -80.5852

# Closest Highway (FL-3/A1A, approximate)
highway_lat = 28.5639
highway_lon = -80.5701


# --- 3. Function to Create Marker and PolyLine ---

def mark_proximity(proximity_lat, proximity_lon, label, color):
    """Calculates distance, creates marker, and draws PolyLine on site_map."""
    
    # Calculate Distance
    distance = calculate_distance(launch_site_lat, launch_site_lon, proximity_lat, proximity_lon)
    
    proximity_coord = [proximity_lat, proximity_lon]
    
    # Create Marker with Distance Label
    distance_marker = folium.Marker(
        proximity_coord,
        icon=DivIcon(
            icon_size=(200, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:{color};"><b>{label}: {distance:10.2f} KM</b></div>',
        )
    )
    site_map.add_child(distance_marker)
    
    # Draw PolyLine
    coordinates = [launch_site_coord, proximity_coord]
    lines = folium.PolyLine(locations=coordinates, weight=2, color=color)
    site_map.add_child(lines)
    
    print(f"Distance to {label}: {distance:.2f} KM")


# --- 4. Execute for each Proximity ---

# Railway
mark_proximity(railway_lat, railway_lon, "Closest Railway", '#00FF00')

# Highway
mark_proximity(highway_lat, highway_lon, "Closest Highway", '#FF0000')

# City
mark_proximity(city_lat, city_lon, "Closest City (Titusville)", '#0000FF')

# Display the final map (The map will now show the launch sites, the success markers, and the distance lines)
site_map
```
